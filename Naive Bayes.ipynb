## Naive Bayes

Naive Bayes is based on the Bayes classification method wherein you compute P(L|Features) from P(Features|L),P(L) & P(Features). It is usually used as a quick method to check if a certain dataset can be classified properly. The 'Naive' part of the name comes from the fact that assumptions should be made for how each of the model was created. That is, to compute P(Features|L) for a label, a 'generative model' is required and assumptions should be made about how this generative model is approximated for each class

```python
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
```

### Gaussian Naive Bayes

In this method, the data is assumed to be from a Gaussian distribution

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, cluster_std=1.5)
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='RdBu')
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y)
```

```python
rng = np.random.RandomState(0)
Xnew = [0,-4] + [7,14] * rng.rand(2000,2)
ynew = model.predict(Xnew)
```

```python
np.unique(ynew)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
plt.axis(lim)
```

```python
yprob = model.predict_proba(Xnew)
yprob[-8:].round(2)
```

### Multinomial Naive Bayes

In this method, the features are assumed to be generated from a multinomial distribution. Multinomial distribution describes the probability of observing counts among a number of categories. It is most appropriate for features that represent counts or count rates. Binomial Distribution is a special form of Multinomial Distribution

```python
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()
data.target_names
```

```python
categories = ['talk.religion.misc', 'soc.religion.christian',
              'sci.space', 'comp.graphics']
train = fetch_20newsgroups(subset='train', categories=categories)
test = fetch_20newsgroups(subset='test', categories=categories)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
model.fit(train.data, train.target)
```

```python
labels = model.predict(test.data)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', 
            cbar=False, xticklabels=train.target_names, 
            yticklabels=train.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
```

```python
def predict_category(s, train=train, model=model):
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
print(train.target_names)
```

```python
predict_category('send payload ISS')
```
